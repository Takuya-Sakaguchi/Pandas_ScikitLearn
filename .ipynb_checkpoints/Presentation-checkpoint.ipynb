## Pandas: series, dataframe and panel

The `Series`: One dimensional dataset, similar to numpy array with special index. It is similar to one column of your excel document.

The `DataFrame`: Two dimensional dataset, equivalent a csv file or one sheet of your excel document. This is the most frequently used data format in Pandas, and we will talk mainly about this format.

The `Panel`: Three dimensional dataset combined multiple dataframes into one format. Conceptually it is similar to an excel document with multiple sheets. This data format is useful but less popular than dataframe.

#  3) Machine lerning with scikit-learn

## Data generation

```python
df.head()
```

## Scikit-learn takes only matrix (numbers), all texts and characters must be comberted to floating points.

```python
def convertSpeciestoNumber(species):
    if species == 'setosa':
        return 0
    elif species == 'versicolor':
        return 1
    elif species == 'virginica':
        return 2
```

```python
convertSpeciestoNumber("setosa")
```

```python
def convertNumbertoSpecies(number):
    if number == 0:
        return 'setosa'
    elif number == 1:
        return 'versicolor'
    elif number == 2:
        return 'virginica'
```

```python
convertNumbertoSpecies(1)
```

## Apply these functions to convert species name to numbers

```python
df["species"]=df['species'].apply(convertSpeciestoNumber)
```

```python
df.tail()
```

## Make train data and label

x_train = train data  
y_train = train label

```python
y_train = df.pop("species")
```

```python
y_train
```

```python
df.head()
```

```python
x_train = df
```

```python
from sklearn.cross_validation import train_test_split
from sklearn import metrics


```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(x_train,y_train,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
Xtrain, Xtest, ytrain, ytest = train_test_split(x_train, y_train, random_state=0)
clf = RandomForestClassifier(n_estimators=100, random_state=0)
clf.fit(Xtrain, ytrain)
ypred = clf.predict(Xtest)
metrics.accuracy_score(ypred, ytest)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
Xtrain, Xtest, ytrain, ytest = train_test_split(x_train,y_train, random_state=0)
gbt = GradientBoostingClassifier(max_depth=9, n_estimators=700)
gbt.fit(Xtrain, ytrain)
ypred = gbt.predict(Xtest)
metrics.accuracy_score(ypred, ytest)
```

```python
newIris = np.array( [ 6.25,  3,  5,  1.5]).reshape(1,-1)
```

```python
clf.predict(newIris)
```

```python
convertNumbertoSpecies(clf.predict(newIris))
```

```python
convertNumbertoSpecies(gbt.predict(newIris))
```
